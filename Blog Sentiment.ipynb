```python
# Function for flagging a specific keyword in the 'Title' column
def keywordflag(keyword):
    flag = []
    length = len(data)
    for h in range(0,length):
        try:
            heading = data['title'][h]
            if keyword in heading:
                flag.append(1)
            else:
                flag.append(0)
        except:
            flag.append(0)
    return flag
```

```python
column_flag = keywordflag('murder')
```

```python
data['Keyword Flag']= pd.Series(column_flag)
```

```python
# Initialization SentimentIntensityAnalyzer
vader = SentimentIntensityAnalyzer()
```

```python
positive_sentiment = []
negative_sentiment = []
neutral_sentiment = []
compound_sentiment = []
text = []
```

```python
lenn = len(data)
```

```python
# Loop for finding out the sentiment scores of each title 
for x in range(0,lenn):
    try:
        text = data['title'][x]
        sent = vader.polarity_scores(text)
        positive_sentiment.append(sent['pos'])
        negative_sentiment.append(sent['neg'])
        neutral_sentiment.append(sent['neu'])
        compound_sentiment.append(sent['compound'])
    except:
        positive_sentiment.append(0)
        negative_sentiment.append(0)
        neutral_sentiment.append(0)
        compound_sentiment.append(0)
```

```python
# Adding the sentiment value columns to the data
data['Positive sentiment'] = pd.Series(positive_sentiment)
data['Negative sentiment'] = pd.Series(negative_sentiment)
data['Neutral sentiment'] = pd.Series(neutral_sentiment)
data['Compound sentiment'] = pd.Series(compound_sentiment)
```
